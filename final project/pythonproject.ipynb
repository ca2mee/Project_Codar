```python
courses = ['Data-Analyst', 'Computer-Science', 'Cyber-Security', 'Empowerment']
```

```python
#Create a small Dataset
for i in range(20):
    student_data= {
    'Student_id': [i for i in range(1, 21)],
    'Name': [fake.name() for _ in range(20)],
    'Age': [random.randint(18, 25) for _ in range(20)],
    'Gender': [random.choice(['Male', 'Female']) for _ in range(20)],
    'Courses': [random.choice(courses) for _ in range(20)],
    'GPA': [round(random.uniform(2.0, 4.0), 2) for _ in range(20)],
    'Study_hours_per_week': [random.randint(10, 40) for _ in range(20)]
}
```

```python
df = pd.DataFrame(student_data)
```

```python
df
```

```python
numerical_attributes = df[['Age', 'GPA', 'Study_hours_per_week']]
```

```python
correlations = numerical_attributes.corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlations, annot=True, cmap='coolwarm', square=True)
plt.title('Correlations between Numerical Attributes')
plt.show()
```

```python
attributes = df[['Age', 'GPA', 'Study_hours_per_week']]
plt.figure(figsize=(10, 8))
sns.pairplot(attributes, markers=['o', 's', 'D'])
plt.title('Pairwise Relationships between Attributes')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='Gender', y='GPA', data=df)
plt.title('GPA Distribution by Gender')
plt.xlabel('Gender')
plt.ylabel('GPA')
plt.show()
```

```python
average_gpa = df['GPA'].mean()
print("Average GPA of all students:", average_gpa)
```
